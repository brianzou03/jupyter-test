# Penguin Classification

```python
import pandas as pd
import seaborn as sns
```

Import necessary libraries - Pandas (handling data), Seaborn (data visualization)

```python
df = pd.read_csv('data/penguin-data.csv').dropna()  # this missing data will cause confusion

# selecting columns rather than passing usecols allows us to reorder them
df[['flipper_length_mm', 'bill_length_mm', 'species']]
```

##### Displaying penguin data on a graph with seaborn

```python
sns.relplot(
    data = df,
    x = 'bill_length_mm',
    y = 'flipper_length_mm',
    hue = 'species',
    height=8,
    hue_order = ['Adelie', 'Gentoo', 'Chinstrap']
)
```
